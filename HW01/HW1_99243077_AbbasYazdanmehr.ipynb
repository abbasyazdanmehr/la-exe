## First: **Finding Span For Vectors**

$
A=\left\lbrack \begin{array}{ccc}
1 & 1 & 1\\
2 & 2 & 2\\
1 & 1 & 1
\end{array}\right\rbrack \Rightarrow \mathrm{Span}\left(S\right)=k_1 \left\lbrack \begin{array}{c}
1\\
2\\
1
\end{array}\right\rbrack +k_2 \left\lbrack \begin{array}{c}
1\\
2\\
1
\end{array}\right\rbrack +k_3 \left\lbrack \begin{array}{c}
1\\
2\\
1
\end{array}\right\rbrack =\left\lbrack \begin{array}{c}
k_1 +k_2 +k_3 \\
2\left(k_1 +k_2 +k_3 \right)\\
k_1 +k_2 +k_3 
\end{array}\right\rbrack =\left\lbrack \begin{array}{c}
q\\
2q\\
q
\end{array}\right\rbrack $

</br>

*description*:  
Matrix A Span(S) is a line in 3d space.

--------

$
B=\left\lbrack \begin{array}{ccc}
1 & 1 & 1\\
2 & -2 & -2\\
1 & -1 & -1
\end{array}\right\rbrack \Rightarrow \mathrm{Span}\left(S\right)=k_1 \left\lbrack \begin{array}{c}
1\\
2\\
1
\end{array}\right\rbrack +k_2 \left\lbrack \begin{array}{c}
1\\
-2\\
-1
\end{array}\right\rbrack +k_3 \left\lbrack \begin{array}{c}
1\\
-2\\
-1
\end{array}\right\rbrack =\left\lbrack \begin{array}{c}
k_1 +\left(k_2 +k_3 \right)\\
2\left(k_1 -\left(k_2 +k_3 \right)\right)\\
k_1 -\left(k_2 +k_3 \right)
\end{array}\right\rbrack =\left\lbrack \begin{array}{c}
q+p\\
2\left(q-p\right)\\
q-p
\end{array}\right\rbrack
$

</br>

*description*:  
Matrix B Span(S) is a plane in 3d space.

-------

$
C=\left\lbrack \begin{array}{ccc}
1 & 1 & -1\\
2 & -2 & 2\\
1 & 1 & 1
\end{array}\right\rbrack \Rightarrow \mathrm{Span}\left(S\right)=k_1 \left\lbrack \begin{array}{c}
1\\
2\\
1
\end{array}\right\rbrack +k_2 \left\lbrack \begin{array}{c}
1\\
-2\\
1
\end{array}\right\rbrack +k_3 \left\lbrack \begin{array}{c}
-1\\
2\\
1
\end{array}\right\rbrack =\left\lbrack \begin{array}{c}
k_1 +k_2 -k_3 \\
2\left(k_1 -k_2 +k_3 \right)\\
k_1 +k_2 +k_3 
\end{array}\right\rbrack
$


</br>

*description*:  
Matrix C Span(S) Contains all 3d space.
